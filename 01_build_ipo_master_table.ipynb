# IPO Success Prediction – Phase 1

## Notebook 01: Build Master IPO Table

**Objective**
- Load official NSE IPO issue data (2010–2024)
- Inspect and validate columns
- Create a clean, time-sorted master IPO table
- No modeling in this notebook

**Data Source**
- NSE IPO Past Issues dataset (2010–2024)

**Research Note**
This notebook strictly focuses on data ingestion and preparation.


```python
import pandas as pd
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
DATA_RAW = PROJECT_ROOT / "data" / "raw"

PROJECT_ROOT, DATA_RAW
```

```python
ipo_master = pd.read_csv(
    DATA_RAW / "IPO-PastIssue-01-01-2010-to-31-12-2024.csv"
)

ipo_master["DATE OF LISTING"] = pd.to_datetime(
    ipo_master["DATE OF LISTING"], errors="coerce", dayfirst=True
)

ipo_master.head()
```

```python
ipo_eq = ipo_master[
    ipo_master["SECURITY TYPE"] == "EQ"
].copy()

ipo_eq = ipo_eq.dropna(subset=["DATE OF LISTING", "Symbol"])

ipo_eq["listing_year"] = ipo_eq["DATE OF LISTING"].dt.year
ipo_eq["listing_month"] = ipo_eq["DATE OF LISTING"].dt.month

ipo_eq.head()
```

```python
out_path = DATA_RAW / "ipo_master_eq.csv"
ipo_eq.to_csv(out_path, index=False)

out_path
```
