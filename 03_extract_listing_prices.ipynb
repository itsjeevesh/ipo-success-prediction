## Extract Listing-Day Prices from NSE Bhavcopy

**Objective**

- Extract listing-day CLOSE price for IPO stocks
- Build a clean price lookup table

**Note**

```python
import pandas as pd
from pathlib import Path
import calendar
```

```python
PROJECT_ROOT = Path.cwd().parent
DATA_RAW = PROJECT_ROOT / "data" / "raw"
BHAV_ROOT = DATA_RAW / "bhavcopy"
IPO_MASTER_PATH = DATA_RAW / "IPO-PastIssue-01-01-2010-to-31-12-2024.csv"
```

```python
ipo = pd.read_csv(IPO_MASTER_PATH)

ipo["DATE OF LISTING"] = pd.to_datetime(
    ipo["DATE OF LISTING"],
    errors="coerce",
    dayfirst=True
)

ipo = ipo.dropna(subset=["DATE OF LISTING", "Symbol"])
ipo = ipo[ipo["SECURITY TYPE"] == "EQ"].copy()

ipo["listing_year"] = ipo["DATE OF LISTING"].dt.year
ipo["listing_month"] = ipo["DATE OF LISTING"].dt.month

ipo.shape
```
